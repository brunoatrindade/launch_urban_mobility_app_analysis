### Testando a hipótese
##### "A duração média dos passeios do bairro Loop para o Aeroporto Internacional O'Hare muda nos sábados chuvosos."

**Formulação das hipóteses**

Quero testar se as durações médias das corridas do Loop até o Aeroporto O’Hare diferem em sábados chuvosos comparadas a sábados sem chuva. Ou seja, quero saber se o clima (chuvoso ou não) influencia o tempo médio das viagens. Dessa forma, formulo minhas hipóteses:

**Hipótese nula (H₀)**:
A duração média das corridas do Loop para O’Hare é a **mesma em sábados chuvosos e sábados sem chuva**.

**Hipótese alternativa (H₁)**:
A duração média das corridas **muda em sábados chuvosos**.

Como vou verificar se há diferença, o teste a ser realizado é bicaudal (two-tailed test). Esse é um teste bicaudal pois não estou assumindo se o tempo aumenta ou diminui com a chuva, apenas quero saber se há diferença significativa.

##### Preparação dos dados e método utilizado

Para testar essa hipótese, vou utilizar o teste t de Student para amostras independentes (scipy.stats.ttest_ind). O critério utilizado se dá em razão de existirem duas amostras independentes, uma de corridas em clima chuvoso (Bad) e outra de corridas não chuvoso (Good). Além disso, as amostras são suficientemente grandes (n > 30), o que torna o teste t indicado mesmo com desvios leves de normalidade.

````python
from scipy import stats

# Carregar os dados
project_sql_7 = pd.read_csv('/datasets/project_sql_result_07.csv')
project_sql_7['start_ts'] = pd.to_datetime(project_sql_7['start_ts'])

# Filtrar apenas os sábados
project_sql_7['day_of_week'] = project_sql_7['start_ts'].dt.day_name()
saturdays = project_sql_7.query("day_of_week == 'Saturday'")

# Separar amostras: chuva (Bad) e sem chuva (Good)
rainy = saturdays.query("weather_conditions == 'Bad'")['duration_seconds']
non_rainy = saturdays.query("weather_conditions == 'Good'")['duration_seconds']

# Teste t de Student (variâncias diferentes)
alpha = 0.05
results = stats.ttest_ind(rainy, non_rainy, equal_var=False)
````

**Critério de decisão**

1. As amostras são independentes, uma corrida em dia chuvoso não afeta outra em dia não chuvoso, sendo nesse caso o teste t apropriado para comparar médias entre dois grupos. O critério de uso teste t de Student para amostras independentes se dá pelo fato de comparação entre duas médias: dias chuvosos e dias não chuvosos.
2. Existem mais de 30 observações em cada grupo, então o Teorema Central do Limite permite o uso do teste t mesmo sem normalidade perfeita.O nível de significância adotado foi α = 0.05, valor padrão em análises de mercado e estudos de mobilidade urbana.

````python
# Exibir resultados
print(f"Média sábados chuvosos: {rainy.mean():.2f}")
print(f"Média sábados não chuvosos: {non_rainy.mean():.2f}")
print(f"Estatística t = {results.statistic:.4f}")
print(f"Valor-p = {results.pvalue:.4f}")

if results.pvalue < alpha:
    print("Rejeitamos H₀: há diferença significativa nas durações.")
else:
    print("Não rejeitamos H₀: não há diferença significativa nas durações.")
````

Em resposta à sugestão do revisor entendi que ele quer a comprovação estatística de qual versão está sendo usada, visando garantir que a escolha da versão do teste-t seja estatisticamente validada, e não apenas presumida. Assim, a análise preliminar da variância entre os grupos (dias chuvosos e dias não chuvosos) é implementada para determinar o parâmetro equal_var e, consequentemente, confirmar a aplicação do Teste de Welch, reforçando a validade e a precisão metodológica do teste de hipótese sobre o impacto da chuva.

````python
# Calcular variâncias das durações em sábados chuvosos e não chuvosos
var_rainy = rainy.var()
var_non_rainy = non_rainy.var()

print(f"Variância sábados chuvosos: {var_rainy:.2f}")
print(f"Variância sábados não chuvosos: {var_non_rainy:.2f}")
````

````python
# Teste formal para igualdade das variâncias (opcional, mas recomendado)
from scipy.stats import levene

stat, p = levene(rainy, non_rainy)
print(f"Estatística de Levene: {stat:.4f}, valor-p: {p:.4f}")

if p < 0.05:
    print("As variâncias são significativamente diferentes: usar equal_var=False (teste de Welch)")
else:
    print("As variâncias são semelhantes: usar equal_var=True (teste t padrão)")
````
